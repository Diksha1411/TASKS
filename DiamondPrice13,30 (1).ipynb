```python
#for dataframe maulpulations and calculations
import pandas as pd
import numpy as np
#for data visualizations
import seaborn as sns
import matplotlib.pyplot as plt
#for knn, linear regression and decision tree algorithms

from sklearn.metrics import mean_squared_error 
from math import sqrt
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import re
from sklearn.ensemble import RandomForestRegressor

#for ignoring warnings
import warnings
warnings.filterwarnings('ignore')

#to get consistent results.
import random
random.seed(0)
```

```python
X=df.drop('price',axis=1)
y=df.price
```

```python
#splitting the data into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 10, test_size = 0.2)
```

```python
#scaling the numerical features for modelling ahead
feat_num=X_train.select_dtypes('number')
mmscaler = MinMaxScaler()
X1= pd.DataFrame(mmscaler.fit_transform(feat_num), columns=feat_num.columns)
X1
```

```python
df.dtypes
```

```python
#scaling the numerical features for modelling ahead
feat_num_test=X_test.select_dtypes('number')
mmscaler = MinMaxScaler()
X1_test= pd.DataFrame(mmscaler.fit_transform(feat_num_test), columns=feat_num_test.columns)
# X1_test
```

```python
#encoding the categorical columns for modelling ahead
feat_cat=X_train.select_dtypes('object')
OHE = OneHotEncoder()
X2=OHE.fit_transform(feat_cat)
X2= pd.DataFrame(OHE.fit_transform(feat_cat).toarray(), columns=OHE.get_feature_names_out())
X2
```

```python
#encoding the categorical columns for modelling ahead
feat_cat_test=X_test.select_dtypes('object')
OHE = OneHotEncoder()
X2_test=OHE.fit_transform(feat_cat_test)
X2_test= pd.DataFrame(OHE.fit_transform(feat_cat_test).toarray(), columns=OHE.get_feature_names_out())
# X2_test
```

```python
#joining the numerical and categorical data 
X_train=pd.concat([X1,X2],axis=1)
X_train.head()
```

```python
#joining the numerical and categorical data 
X_test=pd.concat([X1_test,X2_test],axis=1)
X_test.head()
```

```python
#count of nulls
X_train.isna().sum()
```

```python
#fitting the data
lm = LinearRegression()
lr = lm.fit(X_train,y_train)

R2=lr.score(X_train,y_train)
R2
```

```python
#coefficients and intercept
lr.intercept_, lr.coef_
```

```python
#predicting
yhat_train = lr.predict(X_train)
yhat_test = lr.predict(X_test)
```

```python
#RMSE of train and test
lr_RMSE_train = np.sqrt(sum((y_train-yhat_train)**2)/y_train.shape[0])
lr_RMSE_test = np.sqrt(sum((y_test-yhat_test)**2)/y_test.shape[0])
lr_RMSE_train, lr_RMSE_test
```

```python
#MAPE of train and test
lr_MAPE_train = np.mean((abs(y_train - yhat_train)/y_train)*100)
lr_MAPE_test = np.mean((abs(y_test - yhat_test)/y_test)*100)
lr_MAPE_train,lr_MAPE_test
```

```python
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test, yhat_test)
mean_absolute_percentage_error(y_train, yhat_train)
```
